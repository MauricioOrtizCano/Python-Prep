```python
from bs4 import BeautifulSoup
import urllib.request
```

```python
#En este paso vamos a descargar el archivo Html

reponse = urllib.request.urlopen('https://es.wikipedia.org/wiki/Base_de_datos')
html = reponse.read()
```

```python
soup = BeautifulSoup(html, 'html.parser')
text = soup.get_text()
```
